```python
#18
def calcula_desconto(valor, porcentagem_desconto):
    desconto = valor * (porcentagem_desconto / 100)
    return valor - desconto

valor_produto = float(input("Digite o valor do produto:"))
desconto_aplicado = 15

valor_final = calcula_desconto(valor_produto, desconto_aplicado)
print(f"O produto de R${valor_produto:.2f} com {desconto_aplicado}% de desconto custará R${valor_final:.2f}")
```

```python
#19
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"

nota_aluno = float(input("Digite a nota do aluno:"))
resultado = classifica_nota(nota_aluno)
print(f"Com a nota {nota_aluno}, o aluno está: {resultado}")
```

```python
#20
numeros = [15, -3, 0, 8, -12, 0, 0, 42, -1, 5]

positivos = 0
negativos = 0
zeros = 0

for num in numeros:
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Resumo da lista: {numeros}")
print(f"Números Positivos: {positivos}")
print(f"Números Negativos: {negativos}")
print(f"Exatamente Zero: {zeros}")
```
